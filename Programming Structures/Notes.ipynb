#### Bubble Sort

```python
"""used the hackerearth source(for the visualizer), 
linked somewhere in this document
"""

def bubble_sort(my_array):
    
    for i in range((len(my_array) - 1), 0, -1):
        for n in range(0, len(my_array[0:i])):
            if my_array[n] > my_array[n+1]:
                my_array[n], my_array[n+1] = my_array[n+1], my_array[n]

    return my_array

print(bubble_sort([29, 10, 14, 37, 13]))
```
