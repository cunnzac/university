Quantitative Variables: numbers
1. Discrete: countable |||||
2. Continuous: measurable ~~~~

Qualitative/Categorical Variables: just categories

Sampling Techniques:
1. Cluster: select random clusters of "x" to survey
2. Simple Random: each "x" in population has equal chance of being selected
3. Stratified: divide population into strata (subgroups) based on characteristics, take simple random sample from each, equal representation is goal
4. Systematic: select random seed/starting point, pick every nth "x"
5. Convenience: non-random, sample most convenient "x"
